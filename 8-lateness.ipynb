```python
import numpy as np
import pandas as pd
from bokeh.models import (
    Arrow,
    ColumnDataSource,
    CustomJS,
    CustomJSTickFormatter,
    FixedTicker,
    Grid,
    HoverTool,
    OpenHead,
    WheelZoomTool,
    ColorBar,
    Title
)
from bokeh.events import RangesUpdate, Tap
from bokeh.plotting import figure
from bokeh.transform import dodge

import pipit as pp
from pipit.vis.util import (
    factorize_tuples,
    get_factor_cmap,
    get_html_tooltips,
    get_time_hover_formatter,
    get_time_tick_formatter,
    show,
    trimmed,
)

def lateness_vis(df, coloring_field: str = 'EventName', x_start=None, x_end=None, width=None, height=370):
    from bokeh.transform import linear_cmap
    from bokeh.models import ColumnDataSource, HoverTool
    import bokeh
    from bokeh.plotting import figure, show, output_notebook, curdoc
    from bokeh.models import ColumnDataSource, Range1d
    from bokeh.transform import factor_cmap, dodge
    from bokeh.palettes import Sunset
    import yaml
    from bokeh.themes import Theme

    curdoc().theme = Theme(
        json=yaml.load(
            pp.get_option("theme"),
            Loader=yaml.FullLoader,
        )
    )

    p = figure(sizing_mode="stretch_width" if width is None else "fixed", width=width, height=height, tools="hover", x_axis_label="Logical step", x_axis_location="above") 
    
    source = ColumnDataSource(df)
    p.y_range.flipped = True  # Flip the y-axis
    # index_cmap = factor_cmap('EventName', 
    #                          palette=Spectral5, 
    #                          factors=sorted(df['EventName'].unique()))

    if x_start and x_end:
        p.x_range = Range1d(x_start, x_end)

    
    if coloring_field == 'EventName':
        index_cmap = factor_cmap('EventName', 
                                palette=Sunset, 
                                factors=sorted(df['EventName'].unique()))

    else:
        metric = df[coloring_field]
        # average = metric.mean()
        # std_dev = metric.std()
        # high = average
        # high = average + (std_dev / 16)
        # low = average -  std_dev
        # low = max(low, 0)

        high = df[coloring_field].max()
        low = df[coloring_field].min()
        # pallete = bokeh.palettes.cividis(100)
        # pallete = bokeh.palettes.Plasma256
        # pallete = bokeh.palettes.Turbo256
        pallete = bokeh.palettes.Sunset11
        # pallete = bokeh.palettes.Greys256[::-1]
        index_cmap = linear_cmap(coloring_field, low=low, high=high, palette=pallete)
        # print('min', df[coloring_field].min())
        # print('max', df[coloring_field].max())
        # print('low',low)
        # print('high',high)

    hover = HoverTool(tooltips=[("Process", "@Process"), ("Step", "@Step"), ("Lateness", "@Lateness")])
    p.add_tools(hover)  

    p.hbar(source=source,
        y="Process",
        right=dodge("Step", 0.5),
        left=dodge("Step", -0.5),
        height=0.5,
        line_color="black",
        fill_color=index_cmap,)
        # legend_field="EventName" if coloring_field == "EventName" else "")

    
    # p.yaxis.ticker = bokeh.models.FixedTicker(ticks=list(df['Process'].unique()))
    
    # Additional plot config
    # p.toolbar.active_scroll = p.select(dict(type=WheelZoomTool))[0]

    # Grid config
    process_ticks = np.arange(8)
    p.ygrid.visible = False
    g2 = Grid(
        dimension=1,
        grid_line_width=2,
        band_fill_color="gray",
        band_fill_alpha=0.1,
        ticker=FixedTicker(ticks=process_ticks - 0.5),
        level="underlay",
    )
    p.add_layout(g2)
    # p.segment(source=source,
    #         y0="Process",
    #         x1=dodge("Matching Step", 0),
    #         x0=dodge("Step", 0),
    #         y1="Matching Process", 
    #         color="black")
    
    # Draw arrows
    sends = df[df["EventName"] == "MpiSend"]
    for i in range(len(sends)):
        p.add_layout(
            Arrow(
                end=OpenHead(line_color="black", line_width=1.5, size=8),
                line_color="black",
                line_width=1.5,
                x_start=sends.iloc[i]["Step"],
                y_start=sends.iloc[i]["Process"],
                x_end=sends.iloc[i]["Matching Step"],
                y_end=sends.iloc[i]["Matching Process"], 
                level="overlay",
            )
        )

    # Axis config
    # p.xaxis.formatter = get_time_tick_formatter()
    p.yaxis.formatter = CustomJSTickFormatter(
        code="""
            return "Process " + Math.floor(tick);
        """,
    )
    p.yaxis.ticker = FixedTicker(ticks=process_ticks + 0.1)
    p.yaxis.major_tick_line_color = None

    c = ColorBar(color_mapper=index_cmap.transform, formatter = get_time_tick_formatter(), height=10)
    p.add_layout(c, "below")
    

    show(p) 
```
